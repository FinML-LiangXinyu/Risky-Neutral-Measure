```python
import numpy as np

# 无风险收益率
r = 0.002
# 对数收益率的均值和标准差
mu = 0.01
sigma = 0.3
# 0 时刻的股价
S_0 = 25
# 时间周期 T
T = 250
Times = np.array(range(T))

x = np.random.normal(0, 1, T)
x[0] = 0

# 真实概率测度 P 下的几何布朗运动
P_W_process = x.cumsum()
# 风险中性测度 Q 下的布朗运动
Q_W_process = P_W_process + (mu - r) / sigma * Times
```

```python
import matplotlib.pyplot as plt

# 对比真实概率测度 P 和风险中性测度 Q 下的布朗运动
plt.plot(P_W_process, label = "Real World Probability Measure")
plt.plot(Q_W_process, label = "Risky Neutral Measure")

plt.legend()
```

```python
# 对比真实概率测度 P 和风险中性测度 Q 下的股价随机游走过程
P_S_process = S_0 * np.exp((mu - sigma ** 2 / 2) * sigma * P_W_process)
plt.plot(P_S_process, label = "Real World Probability Measure")
Q_S_process = S_0 * np.exp((r - sigma ** 2 / 2) * sigma * Q_W_process)
plt.plot(Q_S_process, label = "Risky Neutral Measure")

plt.legend()
```
